## üìä Etapa 4: Visualiza√ß√£o dos dados

```python
sns.countplot(x='diagnosis', data=df)
plt.title('Distribui√ß√£o das Classes')
```

```python
df.drop(['id', 'diagnosis', 'Unnamed: 32'], axis=1).hist(bins=15, figsize=(15,10))
plt.tight_layout()
```

```python
# Criar uma nova vers√£o do dataframe s√≥ com colunas num√©ricas
df_numerico = df.drop(['id', 'diagnosis'], axis=1)

plt.figure(figsize=(12,10))
sns.heatmap(df_numerico.corr(), cmap='coolwarm')
plt.title('Correla√ß√£o entre vari√°veis')
```

```python
sns.boxplot(x='diagnosis', y='radius_mean', data=df)
```
